```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
#define vocab
voc_size = 10000
```

```python
one_hot = [one_hot(words,voc_size)for words in sent]
```

```python
one_hot
```

```python
#word embedding 
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
send_len = 8
embeded_docs = pad_sequences(one_hot,padding='pre',maxlen=send_len)
```

```python
embeded_docs
```

```python
#freture representation
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length = send_len))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embeded_docs)
```
